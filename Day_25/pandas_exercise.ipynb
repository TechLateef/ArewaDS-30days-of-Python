Filter the titles which contain python

```python
import re
```

```python
titles_list = []

for title in titles:
    if re.search('python', title, re.IGNORECASE):
        titles_list.append(title)

print(titles_list)        
```

Filter the titles which contain JavaScript

```python
titles_list = []

for title in titles:
    if re.search('javascript', title, re.IGNORECASE):
        titles_list.append(title)

print(titles_list)       
```
